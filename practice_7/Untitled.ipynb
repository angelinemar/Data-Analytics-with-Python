```python
import matplotlib.pyplot as plt 
data = np.random.uniform(0.0, 10.0, 200)
nbin = 5

#print data 
#group data perbin 0-2 interval
print(np.where(data<2))
#print every index mulai dr ke 0, itung total jumlah per index
print(len(np.where(data<2)[0]))
print("-----------------")

for i in range(nbin):
    print(np.where(np.logical_and(data >= 2*i, data < 2*(i+1))))
    print(len(np.where(np.logical_and(data >= 2*i, data < 2*(i+1)))[0]))

plt.hist(data, nbin)
plt.show()
```

```python
Tinggi Badan (X) -> Berat Badan (Y)
160 cm -> 55 kg 
165 cm -> 60 kg 
170 cm -> 65 kg 
175 cm -> 70 kg 
180 cm -> 75 kg

setelah menghitungm kita mungkin medapatkan persamaan regresi linear: 
    
    persamaan: 
        Berat Badan = 30 + 0.25 x Tinggi Badan
    
    Intercept (30): Ini berarti m jika tinggi badan adalah 0 cm 
        (meskipun ini mungkin tidak nyata tinggi badan 0 cm), berat badan diprediksi sekitar 30 kg 
    Slope (0.25): Ini berarti setiap tambahan 1 cm tinggi baan diperkirakan akan meningkatkan berat badan sebesar 0.25 kg
        
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt 

data = {
    'Tinggi Badan (cm)': [160, 165, 170, 175, 100],
    'Berat Badan (kg)': [55, 60, 65, 70, 75]
}

df = pd.DataFrame(data)
print(df)
```

```python
#[[ buat jadiin 2d]]
X = df[['Tinggi Badan (cm)']].values
y = df['Berat Badan (kg)'].values
print(X)
print(y)

X = X.reshape(-1, 1)
print(X)
model = LinearRegression()
model.fit(X, y)

#dapat nilai intercept dan slope 
intercept = model.intercept_
slope = model.coef_[0]

#slope * x + intercept 
print(f'Intercept: {intercept:.2f}')
print(f'Slope: {slope:.2f}')

#prediksi menggunakan model 
# variable y baru 
y_pred = model.predict(X)

#plot data asli (SCATTER)
plt.scatter(df['Tinggi Badan (cm)'], df['Berat Badan (kg)'], color='blue', label='Data Asli')

#plot garis regression (LINE)
plt.plot(df['Tinggi Badan (cm)'], y_pred, color='red', linewidth=2, label='Garis Regresi')

plt.xlabel('Tinggi Badan (cm)')
plt.ylabel('Berat Badan (kg)')
plt.title('Regresi Linear: Berat Badan vs Tinggi Badan')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Menggunakan data car.csv
X = df['Age'].values 
y = df['Speed'].values 
print(X) 
print(y)

#make into 2d array x
X = X.reshape(-1, 1) 
print(X) 
model = LinearRegression()
model.fit(X, y)

#solving intercept and slope
intercept = model.intercept_ 
slope = model.coef_[0]

#f(x) = slope * x + intercept 
print(f'Intercept (0): {intercept:.2f}') 
print(f'Slope (1): {slope:.2f}')

#prediksi menggunakan model 
y_pred = model.predict(X)

#plot data asli 
plt.scatter(df['Age'], df['Speed'], color='blue', label='Data Asli')

#plot garis regresi 
plt.plot(df['Age'], y_pred, color='red', linewidth=2, label='Garis Regresi')

plt.xlabel('Age')
plt.ylabel('Speed')
plt.title('Regresi Linear: Age vs Speed')
plt.legend()
plt.grid(True)
plt.show()
```
