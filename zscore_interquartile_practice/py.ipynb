```python
# 3. Exploratory Data Analysis. Apply Zscore analysis and find the outlier. How many percent(%) outlier data compares to all data? (20%)

data = data[['Front camera']].fillna(data.mean())
data
```

```python
from scipy import stats
import numpy as np

stats.zscore(data)
```

```python
data_baru =data[(np.abs(stats.zscore(data, nan_policy='omit')) < 3).all(axis=1)]
data_baru.shape
```

```python
# Zscore 

data_baru.plot(kind='box', figsize=(5,5))
plt.show

# From the z-score process below, the interquartile range of the car is between 0 - 16.
# Average cars are on 3 - 7 range.
# However, some outliers are indicated to those with significant engine size with the range of 20 - 25.
```

```python
# 4. Data Preprocessing and Preparation. Perform Binning for Prices column with this criteria: (20%)
# a. Below 5 million rupiah categorized as Low
# b. Between 5 million and 10 million categorized as Medium
# c. Above 100 million rupiah categorized as High
```

```python
# Read CSV into pandas again
data = pd.read_csv("onsite-mobile_dataset.csv")
data.head(1000)
```

```python
data.head()
```

```python
# Data Categorization

size = [0, 5000000, 10000000, 100000000]
category = ['Low', 'Medium', 'High']
data['Category'] = pd.cut(data['Price'], bins = size, labels = category)
data
```
