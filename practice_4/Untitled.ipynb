```python
#1. Tampilkan grafik bar rata-rata beer serving di tiap benua

df = pd.read_csv("drinks.csv")
df.head(3)
```

```python
rata = df.groupby('continent')[['beer_servings']].mean().reset_index()
rata
```

```python
rata.plot(x="continent",y="beer_servings", kind="bar")
```

```python
#2. Tampilkan grafik line jumlah negara yang beerserving nya >0 untuk tiap benua

#sort yang beer servingsnya lebih dari 0 dulu 
sort_bs = df[df['beer_servings'] > 0]

#sort 
hasil = sort_bs.groupby('continent')['country'].count()

hasil.plot(x="country", y="continent", legend=True)
```

```python
#3. Tampilkan grafik hbar negara-negara di spirit serving dan beerservingnya diatas 0 untuk setiap benua

sort_ss = df[df['spirit_servings'] > 0]

sort_bs.plot(x="continent", y="beer_servings", kind="barh")
sort_ss.plot(x="continent", y="spirit_servings" ,kind="barh")
```

```python
#3. Tampilkan grafik hbar negara-negara di spirit serving dan beerservingnya diatas 0 untuk setiap benua
benua = df['continent'].unique()
for i in benua:
    print(i)
    sort_ss = df[(df['continent'] == i) & (df['spirit_servings'] > 0) & (df['beer_servings'] > 0)]
    sort_ss.plot(x="country", y="beer_servings", kind="barh")
    sort_ss.plot(x="country", y="spirit_servings" ,kind="barh")
```

```python
# 4. rata-ratakan beerserving, spirit_serving, wine_servings, total_litres_of_pure_alcohol 
# Tampilkan 5 negara yang mempunyai rata-rata tertinggi dalam bentuk barh

df.columns
df['avg_total'] = df['beer_servings']  + df['spirit_servings'] + df['wine_servings'] + df['total_litres_of_pure_alcohol']/4
new = df.sort_values(by="avg_total", ascending=False).head(5)

new.plot(x="country", y="avg_total", kind="barh")
```
