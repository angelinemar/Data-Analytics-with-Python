```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns
df=pd.read_csv('diabetes.csv')
df.head(3)
```

```python
print("Baris", df.shape[0])
print("Kolom", df.shape[1])
print("Daftar Kolom :")
for i in df.columns: 
    print(" - ", i)
```

```python
print("per kolom terdapat berapa baris null")
df.isnull().sum()
```

```python
print("Jumlah cell yang ada null nya: ", sum(df.isnull().sum().values))
```

```python
df.describe()
```

```python
fig, axs = plt.subplots(3,3, figsize=(5,7))
i = 0
for i, col in enumerate(df.columns):
    axs[i//3][i%3].boxplot(df[col])
    axs[i//3][1%3].set_ylabel(col)
plt.show()
```

```python
corr = df.corr()

plt.figure(dpi=130)
sns.heatmap(df.corr(), annot=True, fmt= '.2f')
plt.show()
```
