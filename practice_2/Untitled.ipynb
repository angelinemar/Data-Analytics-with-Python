```python
import pandas as pd
```

```python
df = pd.read_csv("supermarket.csv")
```

```python
# tampilkan 10 data pertama 

df.head(10)
```

```python
# tampilkan 10 data terakhir 

df.tail(10)
```

```python
# tampilkan member yang female 

gb_female = df[df['Gender'] == 'Female']
len(gb_female)
gb_female.head()
```

```python
# tampilkan jumlah member female dan male [groupby]

df.groupby('Gender')['Branch'].count()
```

```python
# Tampilkan Branch dan jumlah customer setiap branch

branch = df[['Branch']]
print(branch)

branch.groupby('Branch')['Branch'].count()
```

```python
# Tampilkan kota secara unik

df['City'].unique()
```

```python
# Tampilkan jumlah kota yang terdapat cabang supermarket ini

df.groupby('Branch')[['City']].count()
```

```python
# Tampilkan jumlah cabang yang dimiliki oleh supermarket ini
df[['Branch']].count()
```

```python
# Tampilkan kota dan setiap kota mempunyai berapa customer

#tampilkan kota 
df['City'].unique()

# tampilkan kota mempunyai berapa cust 
print("setiap kota mempunyai: ")
tes = df.groupby('City')['Customer type'].count()
print(tes)
```

```python
# Tampilkan kota dan setiap kota mempunyai berapa customer hanya untuk gender female saja

female = df[df['Gender'] == 'Female']
female_count = female.groupby('City').size()

print(female_count)
```

```python
# tampilkan kota yang memiliki customer terbanyak 

cust = df.groupby('City')['Invoice ID'].count().reset_index()
cust_max = cust['Invoice ID'].max()

cust[['City']][cust['Invoice ID'] == cust_max]
```

```python
# Tampilkan cabang yang mempunyai customer yang terbanyak

cust = df.groupby('Branch')['Invoice ID'].count().reset_index()
cust.head()

cust_max = cust['Invoice ID'].max()

cust[['Branch']][cust['Invoice ID'] == cust_max]
```

```python
# HELP
# Tampilkan jumlah transaksi setiap bulannya

# convert obj to dt 
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')

# show available months
df['col bulan baru'] = df['Date'].dt.month
df.head(3)

# show list of months 
lists = df.groupby('col bulan baru')[['Invoice ID']].count()
print(lists)
```

```python
# 13. Tampilkan jumlah transaksi setiap bulan percabangnya

jml = df.groupby(['col bulan baru', 'Branch'])[['Invoice ID']].count()
print(jml)
```

```python
# Tampilkan rata-rata quantity product Health and beauty untuk setiap cabangnya
sort = df[df['Product line'] == 'Health and beauty']
sorted = sort.groupby(['Branch','Product line'])[['Quantity']]
sorted.mean()
```

```python
# Tampilkan ada berapa product Electronik accessories yang harganya lebih besar dari 30
df['Product line'].unique()
sort = df[df['Product line'] == 'Electronic accessories']

# len = dapetin jumlah angka
len(sort[sort['Unit price'] > 30])
```

```python
# 16. Tampilkan rata-rata penilaian rating pada product Food and beverages pada setiap kota dan gender

# sort out fnb genre 
sorted_fnb = df[df['Product line'] == 'Food and beverages'].reset_index()
print(sorted_fnb)

# setiap kota
sorted_fnb.groupby('City')[['Rating']].mean()

# setiap gender
sorted_fnb.groupby('Gender')[['Rating']].mean()
```

```python
# 17. Tampilkan berapa customer yang suka produk elektronik yang bergender female yang berada di branch B

sort_elec = df[df['Product line'] == 'Electronic accessories']
sort_elec_2 = sort_elec[sort_elec['Gender'] == 'Female']
sort_elec_3 = sort_elec_2[sort_elec_2['Branch'] == 'B']

sort_elec_3.head(100)
```

```python
# 18. Tampilkan 3 kota yang mementingkan produk kesehatan
df['Product line'].unique()
sort_hnb = df[df['Product line'] == 'Health and beauty']
sort = sort_hnb.groupby(['City', 'Product line'])['Invoice ID'].count().reset_index()

sorted = sort.sort_values(by=['Invoice ID'], ascending=False)
sorted.head(3)
```

```python
# 19. Tampilkan 3 tipe produk favorite dari cabang A 
sort_a = df[df['Branch'] == 'A']
sort = sort_a.groupby('Product line') [['Invoice ID']].count().reset_index()
sort_final = sort.sort_values(by=['Invoice ID'], ascending=False)
sort_final.head(3)
```

```python
# 20. Tampilkan cabang dari kota yang sangat tidak peduli produk kesehatan

sorted = sort.sort_values(by=['Invoice ID'], ascending=True)
sorted.head(3)
```
